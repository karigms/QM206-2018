# Equilibrio#

```python
A=1.0
B=0
k1=0.1
k2=0.2
dt=0.1
t=0
xA=[]
yA=[]
yB=[]

while t<100:
    xA.append(t)
    yA.append(A)
    yB.append(B)
    dA=-k1*A*dt+k2*B*dt
    dB=+K1*A*dt-k2*B*dt
    A+=dA
    B+=dB
    t+=dt
```

```python
plt.plot (xA,yA)
plt.plot(xA,yB)
plt.xlabel("tiempo")
plt.ylabel("concentracion")
plt.title("Equilibrio")
plt.show()
```
