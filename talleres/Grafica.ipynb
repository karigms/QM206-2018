```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([0,2,4,1],[3,1,7,4] ,color="indigo",linewidth=5,alpha=1,marker="o",markersize=20,markeredgewidth=3,markeredgecolor="aquamarine",markerfacecolor="m")
plt.xlabel("$x(cm²)$")
plt.ylabel("$intensidad(\mu)$")
plt.title("Experimento No. 1")
plt.grid(True)
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(4*x)
y2=np.cos(2*x)
plt.plot(x,y,color="indigo")
plt.plot(x,y2,color="crimson",dashes=[4,1,1,2])
plt.title("Función seno(4x)0-$2\pi$")
plt.grid(True)
plt.xlabel("tiempo(seg)")
plt.ylabel("amplitud de la onda")
plt.show()
```

```python
x=np.random.rand(100)*10
y=np.random.rand(100)*10
r=np.random.rand(100)*150+30
data={'a':np.arange(50),'c':np.random.randint(0,50,50),'d':np.random.randn(50)}
data['b']=data['a']+10*np.random.randn(50)
data['d']=np.abs(data['d'])*100
plt.scatter('a','b',c='c',s='d',data=data,alpha=0.8)
plt.show()
```

```python
x=np.random.randn(10000)
plt.hist(x,bins=50,facecolor="indigo",edgecolor="pink",density=True)
plt.title("Distribución de la Población")
plt.ylabel("Densidad de Habitantes/$m²$")
plt.text(1.5,0.35,r"$\mu=0,\ \sigma=1.0$")
plt.show()
```
